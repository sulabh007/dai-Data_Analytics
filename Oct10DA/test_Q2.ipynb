1. Day-wise AQI of India

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

data = pd.read_csv('data_date.csv')
data['Date']=pd.to_datetime(data['Date']).dt.date
india_data = data[data['Country']=='India']
india_data.set_index('Date',inplace=True)

```

```python
day_wise_data = india_data.groupby('Date').mean()
```

```python
plt.plot(day_wise_data.index.tolist(), day_wise_data['AQI Value'],)
```

2. Outliers in AQI of Cyprus

```python
cyprus_data = data[data['Country']=='Cyprus']
cyprus_data.set_index('Date', inplace=True)
plt.boxplot(cyprus_data['AQI Value'],vert=False)
plt.yticks([])
plt.show()
```

```python
#scatter plot
plt.scatter(cyprus_data.index, cyprus_data['AQI Value'], marker='o',color='blue')
plt.title('Cyprus AQI Scatterplot')
plt.xlabel('Date')
plt.ylabel('AQI Value')
plt.grid(True)

#Format x-axis date label
plt.gcf().autofmt_xdate()
plt.tight_layout()
```

3. Outliers in AQI of United Kingdom

```python
uk_data = data[data['Country']=='United Kingdom of Great Britain and Northern Ireland']
uk_data.set_index('Date', inplace=True)
plt.boxplot(uk_data['AQI Value'],vert=False)
plt.yticks([])
plt.show()
```

```python
#scatter plot
plt.scatter(uk_data.index, uk_data['AQI Value'], marker='o',color='blue')
plt.title('United Kingdom AQI Scatterplot')
plt.xlabel('Date')
plt.ylabel('AQI Value')
plt.grid(True)
```

4. Show relation between AQI of India and Nepal

```python
nepal_data = data[data['Country']=='Nepal']
nepal_data.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(10,4),dpi=150)
plt.plot(nepal_data.index.tolist(), nepal_data['AQI Value'],label='Nepal')
plt.plot(india_data.index.tolist(), india_data['AQI Value'],label="India")
plt.legend()
```

```python
#Dendity plot
plt.figure(figsize=(10,6))
sns.kdeplot(data=india_data['AQI Value'],label="India AQI", shade=True)
sns.kdeplot(data=nepal_data['AQI Value'],label="Nepal AQI", shade=True)
plt.title('Density Plot of AQI Values in USA and Canada')
plt.xlabel("AQI Value")
plt.ylabel('Density')
plt.legend()
plt.show()
```

5. Show realtaion between AQI of US and Canada

```python
us_data = data[data['Country']=='United States of America']
us_data.set_index('Date', inplace=True)

canada_data = data[data['Country']=='Canada']
canada_data.set_index('Date', inplace=True)

plt.figure(figsize=(10,4),dpi=150)
plt.plot(us_data.index.tolist(), us_data['AQI Value'],label='USA')
plt.plot(canada_data.index.tolist(), canada_data['AQI Value'],label="Canada")
plt.legend()
```

```python
#Dendity plot
plt.figure(figsize=(10,6))
sns.kdeplot(data=us_data['AQI Value'],label="USA AQI", shade=True)
sns.kdeplot(data=canada_data['AQI Value'],label="Canada AQI", shade=True)
plt.title('Density Plot of AQI Values in USA and Canada')
plt.xlabel("AQI Value")
plt.ylabel('Density')
plt.legend()
plt.show()
```
